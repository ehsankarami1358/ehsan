```python
import numpy as np
from scipy.linalg import svd
from scipy.sparse.linalg import svds
import matplotlib.pyplot as plt
```

```python
R=np.array([[5,3,0,1],[4,0,3,1],[1,1,0,5],[1,0,0,4],[0,1,5,4]],dtype=float)
U,sigma,vt=svds(R,k=2)
#Reconstract
sigma=np.diag(sigma)
predict_matrix=np.dot(np.dot(U,sigma),vt)
print(predict_matrix)
```

```python
#plotting the user embedding in 2d latent space
plt.figure(figsize=(10,10))
plt.scatter(U[:,0],U[:,1])
plt.xlabel('Latent feature 1')
plt.ylabel('Latent feature 2')
plt.title('User embedding in 2D latent space')
for i,text in enumerate(range(U.shape[0])):
    plt.annotate(text,(U[i,0],U[i,1]))
plt.grid()
plt.show()
```
