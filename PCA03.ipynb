```python
from tensorflow.keras.datasets import fashion_mnist
(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.decomposition import PCA
```

```python
x_train_reshaped=X_train.reshape(X_train.shape[0],-1).astype('float32')
x_test_reshaped=X_test.reshape(X_test.shape[0],-1).astype('float32')
min_max_scaler=MinMaxScaler()
x_train_scaled=min_max_scaler.fit_transform(x_train_reshaped)
x_test_scaled=min_max_scaler.transform(x_test_reshaped)
```

```python
pca=PCA(n_components=0.95)
x_train_pca=pca.fit_transform(x_train_scaled)
x_test_pca=pca.transform(x_test_scaled)
```

```python
from sklearn.svm import SVC
svc_linear=SVC(kernel='linear',random_state=42)
svc_linear.fit(x_train_pca,y_train)
```

```python
y_pred=svc_linear.predict(x_test_pca)
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy with PCA:{accuracy}')
```

```python

confusion_matrix=confusion_matrix(y_test,y_pred)
print('Confusion Matrix:')
print(confusion_matrix)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import pandas as pd
```

```python
from ast import Param
param_distribution={
    'C':[0.1,1,10],
    'kernel':['linear','rbf','poly'],
    'gamma':['scale','auto']
                   }
n_combinations=len(param_distribution['C'])*len(param_distribution['kernel'])*len(param_distribution['gamma'])
random_search=RandomizedSearchCV(estimator=svc_linear,param_distributions=param_distribution,n_iter=10,cv=5,random_state=42)
```

```python
results_df=pd.DataFrame({
    'params':list(random_search.cv_results_['params']),
    'Score':random_search.cv_results_['mean_test_score'],
    'Rank':random_search.cv_results_['rank_test_score']
})
results_df.sort_values(by='rank_test_score',inplace=True)
```
