```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.rename(columns={'Annual Income (k$)':'Income','Spending Score (1-100)':'score'},inplace=True)
```

```python
df
```

```python
X=df[['Income','score']]
X
```

```python
X.iloc[:,0]#pd.series
```

```python
X.values[:,0]#numpy array
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X.values[:,0],X.values[:,1],s=15,edgecolors='r',c='k')
plt.xlabel('Income')
plt.ylabel('score')
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X=mms.fit_transform(X)
```

```python
dbscan=DBSCAN(eps=0.1,min_samples=3,metric='euclidean',n_jobs=-1)
clusters=dbscan.fit_predict(X)
clusters
```

```python
for i in range(len(clusters)):
    if clusters[i] !=-1:
        clusters[i]=0
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=clusters,edgecolors='r')
plt.xlabel('Income')
plt.ylabel('score')
```

```python
from sklearn.ensemble import IsolationForest
if_an = IsolationForest(n_estimators=25, max_features=1, max_samples=50, random_state=0, contamination=0.1, warm_start=True) # Changed 'IsolationsForest' to 'IsolationForest'
if_an_list=if_an.fit_predict(X)
```

```python
#if_an_list=pd.Series(if_an_list).replace([-1,-1],[1,0])
#X_if_an=X[if_an_list==1]
```

```python
X_if_an=X[if_an_list==-1]
```

```python
plt.scatter(X[:,0],X[:,1],edgecolors='k',color='w')
plt.scatter(X_if_an[:,0],X_if_an[:,1],edgecolors='r',c='k')
plt.xlabel('Income')
plt.ylabel('score')
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof_an=LocalOutlierFactor(n_neighbors=20,metric='minkowski',p=2,n_jobs=-1)
lof_an_labels=lof_an.fit_predict(X)
```

```python
X_lof_an=X[lof_an_labels==-1]
plt.scatter(X[:,0],X[:,1],edgecolors='k',color='w')
plt.scatter(X_lof_an[:,0],X_lof_an[:,1],edgecolors='r',c='k')
plt.xlabel('Income')
plt.ylabel('score')
```

```python
from sklearn.svm import OneClassSVM
onccsvm_an=OneClassSVM(kernel='rbf',gamma='auto',nu=0.05)
onccsvm_an_labels=onccsvm_an.fit_predict(X)
```

```python
X_ocsvm_an=X[onccsvm_an_labels==-1]
plt.scatter(X[:,0],X[:,1],edgecolors='k',color='w')
plt.scatter(X_ocsvm_an[:,0],X_ocsvm_an[:,1],edgecolors='r',c='k')
plt.xlabel('Income')
plt.ylabel('score')
```

```python
from sklearn.covariance import EllipticEnvelope
ee_en=EllipticEnvelope(random_state=0)
ee_an_labels=ee_en.fit_predict(X)
```

```python
X_ee_an=X[ee_an_labels==-1]
plt.scatter(X[:,0],X[:,1],edgecolors='k',color='w')
plt.scatter(X_ee_an[:,0],X_ee_an[:,1],edgecolors='r',c='k')
plt.xlabel('Income')
plt.ylabel('score')
```
